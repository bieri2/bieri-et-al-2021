# MERRA-2 

```python
#Read in T2M data at the monthly timescale
#Domain corresponds to LPRB
T2M   = get_MERRA2_time_avg('T2M','T2M','M',subset=[-72.0,-40.0,-40.0,-10.0])

#Read in SM data at the monthly timescale
#Domain corresponds to LPRB
SM    = get_MERRA2_time_avg('GWETROOT','SM','M',subset=[-72.0,-40.0,-40.0,-10.0])

#Read in GH500 data at the monthly timescale
GH500 = get_MERRA2_time_avg('H','GH_500mb','M',subset=[-115.0,15.5,-61.5,20.0], fourd=True)

#Read in GH850 data at the monthly timescale
GH850 = get_MERRA2_time_avg('H','GH_850mb','M',subset=[-115.0,15.5,-61.5,20.0], fourd=True)

#Read in pcp data at the monthly timescale
pcp   = get_MERRA2_time_avg('PRECTOTCORR','pcp','M',subset=[-72.0,-40.0,-40.0,-10.0])

#Read in U500 data at the monthly timescale
U500  = get_MERRA2_time_avg('U','UV_500mb','M',subset=[-115.0,15.5,-61.5,20.0],fourd=True)

#Read in V data at the monthly timescale
V500  = get_MERRA2_time_avg('V','UV_500mb','M',subset=[-115.0,15.5,-61.5,20.0],fourd=True)

#Read in LH data at the monthly timescale
LH    = get_MERRA2_time_avg('LHLAND','LH_SH','M',subset=[-72.0,-40.0,-40.0,-10.0])

#Read in SH data at the monthly timescale
SH    = get_MERRA2_time_avg('SHLAND','LH_SH','M',subset=[-72.0,-40.0,-40.0,-10.0])

#Read in U850 data at the monthly timescale
U850  = get_MERRA2_time_avg('U','UV_850mb','M',subset=[-115.0,15.5,-61.5,20.0],fourd=True)

#Read in V850 data at the monthly timescale
V850  = get_MERRA2_time_avg('V','UV_850mb','M',subset=[-115.0,15.5,-61.5,20.0],fourd=True)
```

```python
do_eof(T2M, SM, pcp,'monthly', 1980, 37, filename='meof_merra2_', GH1=GH850, GH2=GH850, U1=U850, V1=V850, U2=U850, V2=V850, LH=LH, SH=SH)
```
